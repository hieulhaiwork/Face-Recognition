```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Khi to FaceDetectorYN
face_detector = cv2.FaceDetectorYN_create(
    model=r"E:\Khac\Face Recognition\Face-Recognition\model\face_detection_yunet_2023mar.onnx",
    config="",
    input_size=(640, 640)
)

# c nh u vo
img = cv2.imread(r"E:\Khac\Face Recognition\Face-Recognition\images\Untitled.png")
img = cv2.resize(img, (640,640))

# Pht hin khun mt
results = face_detector.detect(img)

print(results)

for det in results[1]:
    bbox = det[0:4].astype(np.int32)
    cv2.rectangle(img, (bbox[0], bbox[1]), (bbox[0]+bbox[2], bbox[1]+bbox[3]), (0,255,0), 2)

plt.imshow(img, cmap='gray')
```
